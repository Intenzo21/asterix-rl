```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Functions

```python
# Function for plotting the DDQN training data
def plot_df(dfs, col, x_label, y_label, title, title_fontsize=20, label_fontsize=16):
    plt.rcParams['figure.figsize']= (20, 6)
    for label, df in dfs.items():
        plt.plot(df[col], label=label)
    plt.legend(loc=2, prop={'size': 14})
    plt.xlabel(x_label, fontsize=label_fontsize)
    plt.ylabel(y_label, fontsize=label_fontsize)
    plt.title(title, fontsize=title_fontsize)
    plt.show()
```

```python
def plot_boxplot_for_models(dfs, col, models, x_label, y_label, title, title_fontsize=30, label_fontsize=20):
    fig, ax = plt.subplots(figsize=(30, 10))

    plt.xlabel(x_label, fontsize=label_fontsize)
    plt.ylabel(y_label, fontsize=label_fontsize)
    plt.title(title, fontsize=title_fontsize)
    
    dfs_plot = {}
    for label, df in dfs.items():
        dfs_plot[label] = df[col].values.reshape(-1)
    medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
    meanpointprops = dict(marker='D', markeredgecolor='black', markerfacecolor='firebrick')
    plt.boxplot(dfs_plot.values(), medianprops=medianprops, meanprops=meanpointprops)
    plt.yticks(fontsize=20)
    ax.set_xticklabels(dfs.keys(), fontsize=label_fontsize)

    plt.show()
```

```python
def plot_violinplot_for_models(dfs, col, models, x_label, y_label, title, title_fontsize=30, label_fontsize=20):
    fig, ax = plt.subplots(figsize=(30, 10))

    plt.xlabel(x_label, fontsize=label_fontsize)
    plt.ylabel(y_label, fontsize=label_fontsize)
    plt.title(title, fontsize=title_fontsize)
    
    dfs_plot = {}
    quantiles = []
    for label, df in dfs.items():
        dfs_plot[label] = df[col].values.reshape(-1)
        quantiles.append([0.25, 0.5, 0.75])
    plt.violinplot(dfs_plot.values(), quantiles=quantiles)
    plt.yticks(fontsize=20)
    plt.xticks(ticks=[i for i in range(1, len(dfs.keys())+1)], labels=dfs.keys(), fontsize=label_fontsize)

    plt.show()
```

## Task 1 IMG Plots

```python
data_task_1_dir = './image_agents/results/'
```

```python
data_task_1_1 = pd.read_csv(data_task_1_dir + 'image_1.csv')
```

```python
data_task_1_2 = pd.read_csv(data_task_1_dir + 'image_2.csv')
```

```python
# Plot the training reward cumulative sum throughout the episodes (1 episode = 3 player lives)
plot_df({'Normal Environment': data_task_1_1, 
         'Noop Environment': data_task_1_2}, 
        col=['Total Reward'], 
        x_label='Episode', 
        y_label='Reward', 
        title='DDQN Image-Agent Training Total Reward')
```

```python
# Plot the training reward values throughout the episodes (1 episode = 3 player lives)
plot_df({'Normal Environment': data_task_1_1, 
         'Noop Environment': data_task_1_2}, 
        col=['Total Mean'], 
        x_label='Episode', 
        y_label='Reward', 
        title='DDQN Image-Agent Training Trendline')
```

```python
plot_boxplot_for_models({'Normal Environment': data_task_1_1, 
                         'Noop Environment': data_task_1_2}, 
                        ['Reward'], 
                        ['DDQN-Image'], 
                        x_label='Types of Image-Agents', 
                        y_label='Reward', 
                        title='DDQN Image-Agent Training Rewards BoxPlot for each model')
```

```python
plot_violinplot_for_models({'Normal Environment': data_task_1_1, 
                            'Noop Environment': data_task_1_2}, 
                           ['Reward'], ['DDQN-Image'], 
                           x_label='Types of Image-Agents', 
                           y_label='Reward', 
                           title='DDQN Image-Agent Training Rewards ViolinPlot for each 50,000 steps')
```
